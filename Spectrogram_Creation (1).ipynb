```python
import os
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
from google.colab import drive


# --- CONFIGURATION ---
# The main path to the specific Genre folder
GENRE_PATH = "/content/drive/MyDrive/African Music Genre classifier Dataset/AMGC Dataset/Afrobeat"

# Where the audio files are currently hiding
SOURCE_AUDIO_FOLDER = os.path.join(GENRE_PATH, "Trimmed")

# Where you want the new images to go (Inside the Genre folder, NOT in Trimmed)
OUTPUT_IMAGE_FOLDER = os.path.join(GENRE_PATH, "Spectrograms")

# --------------------------

def create_spectrograms(source_dir, output_dir):
    # Create the output folder if it doesn't exist
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
        print(f"Created new folder: {output_dir}")
    else:
        print(f"Saving to existing folder: {output_dir}")

    # Get list of audio files
    files = [f for f in os.listdir(source_dir) if f.lower().endswith(('.wav', '.mp3'))]
    total_files = len(files)
    print(f"Found {total_files} audio files in {source_dir}...")

    for i, filename in enumerate(files):
        file_path = os.path.join(source_dir, filename)

        try:
            # 1. Load Audio (Force 30s if needed, or just load raw)
            # Using sr=22050 is standard for Mel Spectrograms
            y, sr = librosa.load(file_path, sr=22050)

            # 2. Create Mel Spectrogram
            # n_mels=128 is standard for deep learning models
            S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128, fmax=8000)
            S_db = librosa.power_to_db(S, ref=np.max)

            # 3. Plot (No Axes, No whitespace)
            plt.figure(figsize=(10, 4))
            librosa.display.specshow(S_db, sr=sr, fmax=8000)
            plt.axis('off')  # Turn off axis numbers

            # 4. Save Image
            # Replace .wav with .png
            output_filename = os.path.splitext(filename)[0] + '.png'
            output_path = os.path.join(output_dir, output_filename)

            # Save without borders
            plt.savefig(output_path, bbox_inches='tight', pad_inches=0)
            plt.close() # Close memory to prevent Colab crash

            if (i + 1) % 10 == 0:
                print(f"Processed {i + 1}/{total_files}: {filename}")

        except Exception as e:
            print(f"Error processing {filename}: {e}")

    print(f"\nSuccess! Images saved in: {output_dir}")

# --- RUN IT ---
create_spectrograms(SOURCE_AUDIO_FOLDER, OUTPUT_IMAGE_FOLDER)
```

```python
import os
import csv
import librosa
import numpy as np
import pandas as pd
from google.colab import drive

# 1. Mount Google Drive
if not os.path.exists('/content/drive'):
    drive.mount('/content/drive')

# --- CONFIGURATION ---

# The Main Folder
MAIN_FOLDER = "/content/drive/MyDrive/African Music Genre classifier Dataset/AMGC Dataset"

# The Destination for the CSV (A new folder inside Main)
OUTPUT_FOLDER = os.path.join(MAIN_FOLDER, "Extracted_Features")
OUTPUT_CSV = os.path.join(OUTPUT_FOLDER, "features_extracted.csv")

# Your Specific Genre Paths
# (I mapped the Folder Name to the Label you likely want to use)
GENRE_PATHS = {
    "Afrobeat": "/content/drive/MyDrive/African Music Genre classifier Dataset/AMGC Dataset/Afrobeat/Trimmed",
    "Afrobeats": "/content/drive/MyDrive/African Music Genre classifier Dataset/AMGC Dataset/Afrobeats/Trimmed",
    "Amapiano": "/content/drive/MyDrive/African Music Genre classifier Dataset/AMGC Dataset/Amapiano/Trimmed",
    "Congolese Rumba": "/content/drive/MyDrive/African Music Genre classifier Dataset/AMGC Dataset/Congolese Rumba/Trimmed",
    "Fuji": "/content/drive/MyDrive/African Music Genre classifier Dataset/AMGC Dataset/Fuji/Trimmed",
    "Highlife": "/content/drive/MyDrive/African Music Genre classifier Dataset/AMGC Dataset/Highlife/Trimmed",
    "Juju": "/content/drive/MyDrive/African Music Genre classifier Dataset/AMGC Dataset/Juju/Trimmed",
    "Kwaito": "/content/drive/MyDrive/African Music Genre classifier Dataset/AMGC Dataset/Kwaito/Trimmed",
    "Mbalax": "/content/drive/MyDrive/African Music Genre classifier Dataset/AMGC Dataset/Mbalax/Trimmed",
    "Soukous & Makossa": "/content/drive/MyDrive/African Music Genre classifier Dataset/AMGC Dataset/Soukous & Makossa/Trimmed"
}

# --------------------------

def extract_features_to_csv():
    # 1. Create Output Folder
    if not os.path.exists(OUTPUT_FOLDER):
        os.makedirs(OUTPUT_FOLDER)
        print(f"Created folder: {OUTPUT_FOLDER}")

    # 2. Define the Header (Columns)
    header = 'filename length chroma_stft_mean chroma_stft_var rms_mean rms_var spectral_centroid_mean spectral_centroid_var spectral_bandwidth_mean spectral_bandwidth_var rolloff_mean rolloff_var zero_crossing_rate_mean zero_crossing_rate_var harmony_mean harmony_var perceptr_mean perceptr_var tempo'
    for i in range(1, 21):
        header += f' mfcc{i}_mean mfcc{i}_var'
    header += ' label'
    header = header.split()

    # 3. Create CSV file and write header
    file = open(OUTPUT_CSV, 'w', newline='')
    with file:
        writer = csv.writer(file)
        writer.writerow(header)

    print(f"Created CSV at: {OUTPUT_CSV}")
    print("Starting Feature Extraction... This might take a while.\n")

    # 4. Loop Through Each Genre Folder
    for label, folder_path in GENRE_PATHS.items():
        if not os.path.exists(folder_path):
            print(f"[WARNING] Folder not found: {folder_path} - Skipping.")
            continue

        print(f"--> Processing Genre: {label}")

        # Get all audio files
        files = [f for f in os.listdir(folder_path) if f.lower().endswith(('.wav', '.mp3'))]

        for filename in files:
            song_path = os.path.join(folder_path, filename)

            try:
                # Load Audio (Preserve native sample rate or force 22050)
                y, sr = librosa.load(song_path, mono=True, sr=22050)

                # --- EXTRACT FEATURES ---

                # Length
                length = len(y)

                # STFT based features
                chroma_stft = librosa.feature.chroma_stft(y=y, sr=sr)
                rms = librosa.feature.rms(y=y)
                spec_cent = librosa.feature.spectral_centroid(y=y, sr=sr)
                spec_bw = librosa.feature.spectral_bandwidth(y=y, sr=sr)
                rolloff = librosa.feature.spectral_rolloff(y=y, sr=sr)
                zcr = librosa.feature.zero_crossing_rate(y)

                # Harmony & Percussive
                y_harm, y_perc = librosa.effects.hpss(y)
                harmony_mean = np.mean(y_harm)
                harmony_var = np.var(y_harm)
                perceptr_mean = np.mean(y_perc)
                perceptr_var = np.var(y_perc)

                # Tempo
                tempo, _ = librosa.beat.beat_track(y=y, sr=sr)

                # MFCCs (1-20)
                mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=20)

                # --- ASSEMBLE ROW ---
                to_append = f'{filename} {length}'
                to_append += f' {np.mean(chroma_stft)} {np.var(chroma_stft)}'
                to_append += f' {np.mean(rms)} {np.var(rms)}'
                to_append += f' {np.mean(spec_cent)} {np.var(spec_cent)}'
                to_append += f' {np.mean(spec_bw)} {np.var(spec_bw)}'
                to_append += f' {np.mean(rolloff)} {np.var(rolloff)}'
                to_append += f' {np.mean(zcr)} {np.var(zcr)}'
                to_append += f' {harmony_mean} {harmony_var}'
                to_append += f' {perceptr_mean} {perceptr_var}'
                to_append += f' {tempo}'

                for e in mfcc:
                    to_append += f' {np.mean(e)} {np.var(e)}'

                # Add Label at the end
                to_append += f' {label}'

                # Write to CSV
                file = open(OUTPUT_CSV, 'a', newline='')
                with file:
                    writer = csv.writer(file)
                    writer.writerow(to_append.split())

            except Exception as e:
                print(f"[ERROR] Could not process {filename}: {e}")

    print(f"\nSUCCESS! All features saved to: {OUTPUT_CSV}")

# --- RUN THE FUNCTION ---
extract_features_to_csv()
```
